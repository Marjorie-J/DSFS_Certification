# 2. Comparaison Clustering / Thématiques

```python
df_corpus["cluster"].value_counts()
```

```python
df_corpus["main_topic"].value_counts()
```

```python
# Création de la table croisée
contingency_table = pd.crosstab(df_corpus["cluster"], df_corpus["main_topic"])
contingency_table
```

# 3. Conclusion

Les thématiques identifiées par LSA et les clusters issus du clustering sont différents mais en même temps ils se recoupent. 

En effet, contrairement au clustering qui associe chaque document à un seul groupe, le LSA permet d’attribuer à chaque document plusieurs thématiques, reflétant ainsi la complexité et la richesse des contenus.

Cette complémentarité souligne l’intérêt de combiner ces approches : le clustering apporte une vision segmentée du corpus, tandis que le LSA permet une compréhension plus nuancée des relations thématiques entre documents.
