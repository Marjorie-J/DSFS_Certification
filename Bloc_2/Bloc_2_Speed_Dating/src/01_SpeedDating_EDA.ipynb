## Analyse des participants : Qui sont-ils ?

#### Age

```python
fig = px.histogram(df_sd, x="age", color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_layout(yaxis_title="Nombre d'individus", title="Distribution des âges", title_x=0.5)
fig.show()
```

La distribution d'âge est-elle différente pour les hommes et les femmes ?

```python
df_sd['gender_label'] = df_sd['gender'].map({0: 'Femme', 1: 'Homme'})

fig = px.histogram(df_sd, x="age", color="gender_label", labels={"gender_label": "Genre", "age": "Âge"}, color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_layout(yaxis_title="Nombre d'individus", title="Distribution des âges par genre", title_x=0.5)
fig.show()
```

```python
fig = px.box(df_sd, x="gender_label", y = "age", labels={"gender_label": "Genre", "age": "Âge"}, color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_layout(title="Distribution des âges par genre", title_x=0.5)
fig.show()
```

La population semble être composée en majorité de jeunes adultes (environ 20–30 ans).

Nous n'observons pas de grosse différence d'âge selon le genre.
