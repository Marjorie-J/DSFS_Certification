#### Je choisi ensuite d'explorer les donnÃ©es sous l'axe suivant : A quels attributs les participants accordent-ils le plus dâ€™importance dans la dÃ©cision de revoir un partenaire ?

```python
# SÃ©lection des colonnes pertinentes pour l'analyse : dÃ©cision de revoir le partenaire + Ã©valuations (de 1 Ã  10) des partenaires sur la base de 6 attributs
columns_scorecard = [
    "dec",
    "attr",
    "sinc",
    "intel",
    "fun",
    "amb",
    "shar",
]

# VÃ©rifier les valeurs manquantes
missing_values = df_sd[columns_scorecard].isnull().sum()
print(missing_values)
```

Observations sur les valeurs manquantes :
- dec : Aucune valeur manquante
- attr contient des valeurs manquants : 2.4% du dataset
- sinc contient des valeurs manquants : 3.3% du dataset
- intel contient des valeurs manquants : 3.5% du dataset
- fun contient des valeurs manquants : 4.1% du dataset
- shar contient des valeurs manquants : 12.7% du dataset

Je nettoye les donnÃ©es en supprimant les valeurs manquantes avant d'explorer les relations entre ces attributs et la dÃ©cision de revoir le partenaire. â€‹

```python
# Suppression des lignes avec des valeurs manquantes
df_sd_clean = df_sd[columns_scorecard].dropna()

# VÃ©rifier que les valeurs manquantes ont Ã©tÃ© supprimÃ©es
df_sd_clean.isnull().sum()
```

Le dataset est propre

```python
# On remplace la dÃ©cision de revoir par "Oui" ou "Non", pour la visualisation
df_sd_clean["dec"] = df_sd_clean["dec"].apply(lambda x: "Oui" if x == 1 else "Non")
display(df_sd_clean)
```

Nous allons analyser la moyenne de la notation de chaque attribut selon la dÃ©cision de revoir

```python
# Moyenne de la notation de chaque attribut selon la dÃ©cision de revoir
df_mean = df_sd_clean.groupby("dec")[["attr", "sinc", "intel", "fun", "amb", "shar"]].mean().reset_index()
display(df_mean)
```

```python
# on retravaille le dataframe pour avoir, pour chaque dÃ©cision, l'attribut, la moyenne des notes pour cet attribut
melted_df = df_mean.melt(id_vars=["dec"], value_vars=["attr", "sinc", "intel", "fun", "amb", "shar"])
melted_df["variable"] = melted_df["variable"].apply(lambda x: "AttractivitÃ©" if x == "attr" 
                                                    else "SincÃ©ritÃ©" if x == "sinc"
                                                    else "Intelligence" if x == "intel"
                                                    else "Humour" if x == "fun"
                                                    else "Ambition" if x == "amb"
                                                    else "IntÃ©rÃªts communs" if x == "shar" else "Aucun")
display(melted_df)
```

```python
#Visualisation des moyennes des notes des attributs selon la dÃ©cision prise
fig = px.bar(melted_df, x="variable", y="value", color="dec", barmode="group", title="Moyenne des attributs selon la dÃ©cision de revoir")
fig.update_layout(title_x=0.5, xaxis_title="Attributs",  yaxis_title="Valeur Moyenne", legend_title="DÃ©cision de revoir")
```


L'attractivitÃ©, lâ€™humour et les intÃ©rÃªts communs ont des Ã©carts plus marquÃ©s entre ceux qui sont souhaitÃ©s Ãªtre revus et ceux qui ne le sont pas.

Cela pourrait indiquer que ces critÃ¨res sont particuliÃ¨rement importants dans la prise de dÃ©cision contrairement Ã  la sincÃ©ritÃ©, l'intelligence et l'ambition.

Nous allons donc analyser plus prÃ©cisÃ©mment la distribution des notes pour les 3 attributs : attractivitÃ©, humour et intÃ©rÃªts communs.

```python
#Distribution des attributs AttractivitÃ©, Humour et IntÃ©rÃªts communs selon la dÃ©cision

attributes = {"attr" : "AttractivitÃ©", "fun" : "Humour", "shar" : "IntÃ©rÃªts communs"}

# CrÃ©ation des subplots
fig = make_subplots(rows = 1, cols = 3, subplot_titles=[attributes[attr] for attr in attributes])

mask_oui = (df_sd_clean["dec"] == "Oui")
mask_non = (df_sd_clean["dec"] == "Non")

# Ajout des boxplots dans les subplots
for i, attr in enumerate(attributes):
    # DÃ©cision Oui
    fig.add_trace(
        go.Box(y=df_sd_clean[mask_oui][attr], name="Oui", marker_color="red"),
        row=1, col=i+1
    )
    # DÃ©cision Non
    fig.add_trace(
        go.Box(y=df_sd_clean[mask_non][attr], name="Non", marker_color="blue"),
        row=1, col=i+1
    )

fig.update_layout(title="Distribution des notes des attributs selon la dÃ©cision", title_x=0.5, height=500, width=1200, showlegend=False)
fig.show()
```

Observations :

1. AttractivitÃ©
"Oui" a une mÃ©diane plus Ã©levÃ©e (7) que "Non" (5).

Lâ€™Ã©tendue des scores "Oui" est plus concentrÃ©e (de 3 Ã  10, avec quelques valeurs aberrantes basses), tandis que "Non" sâ€™Ã©tend plus largement (de 0 Ã  10).

Les "Oui" semblent avoir une tendance gÃ©nÃ©rale Ã  Ãªtre jugÃ©s plus attractifs.

2. Humour

Les valeurs extrÃªmes (outliers) apparaissent des deux cÃ´tÃ©s.

3. IntÃ©rÃªts communs
MÃ©diane plus Ã©levÃ©e pour "Oui" (7) par rapport Ã  "Non" (5).

Le groupe "Non" a une Ã©tendue complÃ¨te (0 Ã  10), alors que "Oui" est plus concentrÃ© entre environ 1 et 10.

Peu ou pas dâ€™outliers visibles ici.

InterprÃ©tation gÃ©nÃ©rale :

Pour AttractivitÃ© et IntÃ©rÃªts communs, le groupe "Oui" tend Ã  avoir des valeurs plus Ã©levÃ©es et plus concentrÃ©es.

Pour Humour, la diffÃ©rence est moins marquÃ©e mais "Oui" reste un peu au-dessus en mÃ©diane.

Les outliers sont plus frÃ©quents dans le groupe "Non", ce qui peut indiquer plus de variabilitÃ© dans les notes donnÃ©es Ã  ce groupe.




- Peu d'outliers (valeurs extrÃªmes). Peu de personnes ont reÃ§u des notes trÃ¨s diffÃ©rentes de la majoritÃ©

- Quartiles :

```python
# SÃ©parer les vagues selon le systÃ¨me de notation
waves_6_9 = df_sd[(df_sd['wave'] >= 6) & (df_sd['wave'] <= 9)]
waves_other = df_sd[(df_sd['wave'] <= 5) | (df_sd['wave'] >= 10)]
```

```python
attributes = ["attr1_1", "sinc1_1", "intel1_1", "fun1_1", "amb1_1", "shar1_1"]
```

```python
# --- Analyse pour waves 6-9 (notes 1-10) ---
mean_attr_6_9 = waves_6_9.groupby('gender_label')[attributes].mean()
```

```python
mean_attr_6_9.rename(columns={
    'attr1_1': 'AttractivitÃ©',
    'sinc1_1': 'SincÃ©ritÃ©',
    'intel1_1': 'Intelligence',
    'fun1_1': 'Humour',
    'amb1_1': 'Ambition',
    'shar1_1': 'IntÃ©rÃªts communs'
}, inplace=True)
```

```python
# --- Analyse pour waves 1-5 & 10-21 (rÃ©partition 100 points) ---
mean_attr_other = waves_other.groupby('gender_label')[attributes].mean()
```

```python
mean_attr_other.rename(columns={
    'attr1_1': 'AttractivitÃ©',
    'sinc1_1': 'SincÃ©ritÃ©',
    'intel1_1': 'Intelligence',
    'fun1_1': 'Humour',
    'amb1_1': 'Ambition',
    'shar1_1': 'IntÃ©rÃªts communs'
}, inplace=True)
```

```python
# HOMMES

top3_homme_6_9 = (
    mean_attr_6_9.loc['Homme']
    .sort_values()
    .head(3)
    .reset_index()
    .rename(columns={'index': 'attribute', 'Homme': 'importance'})
)
top3_homme_6_9['method'] = 'Waves 6â€“9 : Ã‰chelle 1â€“10'

top3_homme_other = (
    mean_attr_other.loc['Homme']
    .sort_values()
    .head(3)
    .reset_index()
    .rename(columns={'index': 'attribute', 'Homme': 'importance'})
)
top3_homme_other['method'] = 'Waves 1â€“5 & 10â€“21 : 100 points'

top3_homme_combined = pd.concat([top3_homme_6_9, top3_homme_other])
```

```python
# FEMMES

top3_femme_6_9 = (
    mean_attr_6_9.loc['Femme']
    .sort_values()
    .head(3)
    .reset_index()
    .rename(columns={'index': 'attribute', 'Femme': 'importance'})
)
top3_femme_6_9['method'] = 'Waves 6â€“9 : Ã‰chelle 1â€“10'

top3_femme_other = (
    mean_attr_other.loc['Femme']
    .sort_values()
    .head(3)
    .reset_index()
    .rename(columns={'index': 'attribute', 'Femme': 'importance'})
)
top3_femme_other['method'] = 'Waves 1â€“5 & 10â€“21 : 100 points'

top3_femme_combined = pd.concat([top3_femme_6_9, top3_femme_other])
```

```python
# CrÃ©er les subplots (2 ligne, 2 colonnes)
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=(
        "Hommes",
        "Femmes"
    )
)

# Hommes : Waves 6-9
fig.add_trace(
    go.Bar(
        x=top3_homme_6_9['attribute'],
        y=top3_homme_6_9['importance'],
        name="Homme",
        marker_color='steelblue',
        text=top3_homme_6_9['importance'].round(2)
    ),
    row=1, col=1
)

# Hommes : Waves Autres
fig.add_trace(
    go.Bar(
        x=top3_homme_other['attribute'],
        y=top3_homme_other['importance'],
        name="Homme",
        marker_color='steelblue',
        text=top3_homme_other['importance'].round(2)
    ),
    row=2, col=1
)

# Femmes : Waves 6-9
fig.add_trace(
    go.Bar(
        x=top3_femme_6_9['attribute'],
        y=top3_femme_6_9['importance'],
        name="Homme",
        marker_color='indianred',
        text=top3_femme_6_9['importance'].round(2)
    ),
    row=1, col=2
)

# Femmes : Waves Autres
fig.add_trace(
    go.Bar(
        x=top3_femme_other['attribute'],
        y=top3_femme_other['importance'],
        name="Homme",
        marker_color='indianred',
        text=top3_femme_other['importance'].round(2)
    ),
    row=2, col=2
)

# Layout final
fig.update_layout(
    title_text="Top 3 attributs les moins dÃ©sirables, par genre, par systÃ¨me de notation", 
    showlegend=False
)
fig.update_yaxes(title_text="Waves 6-9", row=1)
fig.update_yaxes(title_text="Waves 1-5, 10-21", row=2)
fig.show()
```
