Pour chaque ville :
 - Récupère les coordonnées géographiques (latitude et longitude), grâce à l'API https://nominatim.org/ 
 - Rcupère les informations météorologiques pour les 5 jours à venir, grâce à l'API https://openweathermap.org/

```python
import pandas as pd
import requests
from variables import best_cities, cities_csv_name, weather_csv_name

cities = []
weathers = []

city = {}
weather = {}

for i in range(len(best_cities)): 

    city = {}
    city["id"] = i + 1
    city["city"] = best_cities[i]

    print("Get weather information", city["city"])

    # Paramètre de l'API l'API https://nominatim.org/ 
    params_geo = {"format": "json", 
                "limit": "1",
                "email": "janda.marjorie@gmail.com",
                "city": city["city"]}

    url_geo = f"https://nominatim.openstreetmap.org/search"

    # Paramètres de l'API https://openweathermap.org/
    params_weather = {"appid": "061dd2f47cb411e9cfecdd4437d13f6a", 
                    "units": "metric",
                    "lang": "fr",
                    "mode": "json"}

    # Ici le endpoint appelé est le endpoint forecast qui retourne une météo sur 5 jours
    # Le endpoint onecall 2.5, qui retourne une météo sur 7 jours, est déprecié
    # Un onecall 3.0 a été mis en place, mais payant
    url_weather = "https://api.openweathermap.org/data/2.5/forecast"

    #Appel API géolocalisation
    try:
        response_geo = requests.get(url_geo, params_geo)
    
        #API géolocalisation OK
        if response_geo.status_code == 200:

            if response_geo.json():
                data_geo = response_geo.json()[0]
                
                city["lat"] = float(data_geo["lat"])
                city["lon"] = float(data_geo["lon"])

                params_weather["lat"] = data_geo["lat"]
                params_weather["lon"] = data_geo["lon"]
                
                #Appel API météo
                try:
                    response_weather = requests.get(url_weather, params_weather)
                
                    #API météo OK
                    if response_weather.status_code == 200:
                        data_weather = response_weather.json()["list"]

                        #Lecture des données de l'API météo +
                        #Alimentation des listes pour peupler le fichier csv
                        for d_weather in data_weather:
                            weather = {}
                            weather["id_city"] = city["id"]
                            weather["date"] = d_weather["dt_txt"]
                            weather["temperature"] = d_weather["main"]["temp"]
                            weather["temperature_feels_like"] = d_weather["main"]["feels_like"]
                            weather["condition"] = d_weather["weather"][0]["description"]
                            weather["wind_speed"] = d_weather["wind"]["speed"]
                            weather["proba_precipitation"] = int(d_weather["pop"] * 100)
                            
                            weathers.append(weather)

                    else:
                        print("Status Weather Error:", response_weather.status_code)

                except requests.exceptions.RequestException as error:
                    print("Exception Weather Error:", error)

        else:
            print("Status Geo Error:", response_geo.status_code)

    except requests.exceptions.RequestException as error:
        print("Exception Geo Error:", error)

    cities.append(city)

print("Ecriture CSV")

# Ecriture des listes en csv

df_city = pd.DataFrame(cities) 
df_city.to_csv(cities_csv_name, index=False)

df_weather = pd.DataFrame(weathers) 
df_weather.to_csv(weather_csv_name, index=False)
```
